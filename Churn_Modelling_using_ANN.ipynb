**Data Preprocessing**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Dataset => https://www.kaggle.com/datasets/shrutimechlearn/churn-modelling
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df
```

```python
X = df.iloc[:,3:13].values
Y = df.iloc[:,13].values
```

**Label encoding using scikit-learn**:
Geography & Gender are not numerical we need to transfer them this is encoding for eg. female : 1, male: 0
For geography column : hot encoding will use

```python
# gender ko 0 & 1 me convert kiya of X
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_2 = LabelEncoder()
X[:,2]=labelencoder_X_2.fit_transform(X[:,2])
```

**One Hot Encoding**

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('ohe',OneHotEncoder(),[1])],remainder='passthrough')
X = np.array(ct.fit_transform(X),dtype=str) #.toarray()
X = X[:,1:]
```

```python
print(X)
```

```python
# splitting in train & test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, random_state=0)
```

**Feature scaling helps to normalize the data within a specific range, ensuring that each feature contributes proportionately to the result.**

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
X_train = sc.fit_transform(X_train) # we learn the param of scaling on training data
X_test = sc.transform(X_test) # we use the scaling params learned on training data to scale the test data
```

```python
import pandas as pd
print(X_train)
```

```python
pd.DataFrame(X_train)
```

Building the ANN

```python
from tensorflow.keras.models import Sequential
from keras.models import Sequential # Sequential model/class
from keras.layers import Dense # Dense layer
classifier = Sequential() # object
```

Adding input layer & first hidden layer

```python
classifier.add(Dense(units=6, kernel_initializer ='uniform',activation='relu',input_dim =11))
# add layers to our model 11 input neurons & 6 output neurons
# units = Last layer neurons, kernel_initialiser = func to initialise weights for uniform data is uniformly distributed,
# rectifier activation function , input_dim = 11 why 11 => 11 independent variables
```

Adding next hidden layer

```python
classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu'))
# no input_dim because it automatically gets connected to previous layer
```

Adding output layer : we need only 1 neuron

```python
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
# 1 neuron is enough to predict output
# sigmoid is used to predict whether the customer will stay or not in the bank
```

Compiling the ANN

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
#compile = method of tensorflow library
# adam => uses stochastic gradient descent technique
# binary_crossentropy => loss func (optimization func which is used in case of training a classification model
# which classifies which tells whether the data belongs to one class/other)
# evaluation of network => accuracy metric which creates 2 local variable
# total & count => compute frequency matches why_pred matches with why_true
```

Fitting the ANN to training set

```python
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
# 1st arg => input neurons
# 2nd arg => output layer neurons
# batch size = perform in batch
# epochs => number of time algo visits the dataset
```

Predictiong the test set

```python
y_pred = classifier.predict(X_test)
y_pred =(y_pred>0.5)
# if y_pred is between 0 & 0.5 the new y_pred will become zero
```

```python
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
accuracy_score(y_test,y_pred)
```
